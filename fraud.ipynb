```python
encoder = LabelEncoder()
columns_to_encode = ['merchant', 'category', 'street', 'job', 'trans_num', 'first', 'city', 'state', 'last', 'gender', 'trans_date_trans_time', 'dob']

for column in columns_to_encode:
     fraud_train[column] = encoder.fit_transform(fraud_train[column])
        
```

```python
for column in columns_to_encode:
    fraud_test[column] = encoder.fit_transform(fraud_test[column])
```

```python
x_train=fraud_train.drop(columns='is_fraud')
y_train=fraud_train['is_fraud']
x_test=fraud_test.drop(columns='is_fraud')
y_test=fraud_test['is_fraud']
```

```python

model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
predictions=model.predict(x_test)
```

```python

cm=confusion_matrix(y_test,predictions)

cm
```

```python
accuracy=accuracy_score(y_test,predictions)
accuracy
```
